```python
import random
import datetime
def get_noise_for_houses(df: pd.DataFrame, num_samples: int) -> pd.DataFrame:
  """
  Добавляет num_samples образцов с одной отличающейся характеристикой (больше 3сигм)
  """
  data_to_include = []
  stats = df.describe()
  last_id = int(stats.loc['max']['id'])+1
  colnames = list(df.columns)
  colnames.remove('id')
  colnames.remove('date')
  for index in range(last_id, last_id+num_samples):
    distorshion_scaler = random.randint(10, 20)
    random_column_name = random.choice(colnames)
    copy_of_sample = stats.loc['50%']+distorshion_scaler*stats.loc['std']
    copy_of_sample['date'] = datetime.datetime.strptime('2015-02-23', '%Y-%m-%d')
    copy_of_sample['id'] = index
    # copy_of_sample[random_column_name] = stats.loc['mean'][random_column_name]+distorshion_scaler*stats.loc['std'][random_column_name]
    data_to_include.append(copy_of_sample)
    # print(random_column_name)

  return pd.DataFrame(data=data_to_include)

num_anomalies = 1500
noisy = get_noise_for_houses(house_data, num_anomalies)
house_anomaly = pd.concat([house_data, noisy]).reset_index(drop=True)
house_anomaly = house_anomaly.drop(columns=['id', 'date'])
```

```python
house_anomaly.tail(num_anomalies)
```

```python
house_anomalies_indexes = house_anomaly.index.values[-num_anomalies:]
house_anomalies_indexes
```

```python
credit_anomaly_indexes = credit_anomaly[credit_anomaly['Class']==1].index.values
credit_anomaly_indexes.shape
```

```python
credit_anomaly = credit_anomaly.drop(columns='Class')
credit_anomaly
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.neighbors import LocalOutlierFactor
import matplotlib.pyplot as plt
import numpy as np

plots_per_row = 3
subset=-1
datasets = {'houses': [house_anomaly.iloc[:subset], house_anomalies_indexes],
            'transactions': [credit_anomaly.iloc[:subset], credit_anomaly_indexes]}
methods = {
    'IsolationForest':{
        'model': IsolationForest(n_jobs=-1),
        'houses': {'max_features':1, 'max_samples': 1.0},
        'transactions': {'max_features':5, 'max_samples': 1.0}},
    'LocalOutlierFactor': {
        'model': LocalOutlierFactor(n_jobs=-1),
        'houses': {'n_neighbors':1000},
        'transactions': {'n_neighbors': 1000}}
    }

def plot_distribution(values: list, anomaly_indeces: list, other: list, ax:plt.Axes, feature_name: str):
  # plt.figure()
  ax.set_yscale('log')
  ax.hist(values, bins=len(values), histtype='step')
  ax.hist(values[anomaly_indeces], histtype='step', bins=len(anomaly_indeces), color='red')
  ax.set_title(feature_name)

for dataset_name, dataset in list(datasets.items()):
  print(dataset_name)
  for method, args in list(methods.items()):
    print(method)

    if method in ['IsolationForest', 'LocalOutlierFactor']:
      args[dataset_name]['contamination'] = len(dataset[1])/len(dataset[0])
    else:
      args[dataset_name]['nu'] = len(dataset[1])/len(dataset[0])

    model = args['model']
    predictions = model.set_params(**args[dataset_name]).fit_predict(dataset[0])
    other = np.argwhere(predictions==1).reshape(1,-1)[0]
    anomalies = np.argwhere(predictions==-1).reshape(1,-1)[0]

    recall = len(set(anomalies).intersection(set(dataset[1])))/len(dataset[1])

    n_features = len(dataset[0].columns)
    figure, axes = plt.subplots(nrows=n_features//plots_per_row+1, ncols=plots_per_row)
    figure.suptitle(f'Total samples in {dataset_name} is {len(dataset[0])}'
              f'\nTotal anomalies detected with {method} is {len(anomalies)}\n'
              f'Actual number of anomalies is {len(dataset[1])}. Recall is {recall}')

    figure.set_figwidth(10)
    figure.set_figheight(15)

    for feature_name, ax in list(zip(dataset[0].columns, axes.ravel())):
      plot_distribution(dataset[0][feature_name], anomalies, other, ax, feature_name)

    plt.tight_layout()
    plt.show()

# for feature
```
